```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import os
dir=[i for i in os.listdir('Samsung_Train_Image/Train')]
```

In this work we train our system with different samsung logo  and mark the  samsung logo on the samsung banner. The thresehold accuracy of the system is 0.8.

```python
a=[]
gr=[]
import numpy as np
maxl=0
index=0
img_bgr=[]
loc=np.array([])
locl=np.array([])
w=0
h=0
for h in range(0,len(dir)):
    img_bgr = cv2.imread('Samsung_Train_Image/Test/samsung3.jpeg')
    img_gray = cv2.imread('Samsung_Train_Image/Test/samsung3.jpeg',0)
 #   print  "Image size "+str(img_gray.shape)
    a.append(cv2.imread('Samsung_Train_Image/Train/'+dir[h]))
   # gr.append(cv2.cvtColor(a[h], cv2.COLOR_BGR2GRAY))
    gr.append(cv2.imread('Samsung_Train_Image/Train/'+dir[h]))
    temp=cv2.imread('Samsung_Train_Image/Train/'+dir[h],0)
  #  print  "Temp size "+str(temp.shape)
    #print "Res "+dir[h] +" "+str(res)
    res=cv2.matchTemplate(img_gray,temp,cv2.TM_CCOEFF_NORMED)
    
    thrsehold=0.8
    loc=np.where(res>=thrsehold)
    #print loc.sum()
    #if locl.size < loc.size:
    #    locl=loc
    #print dir[h]
    if len(loc[0])>maxl:
        maxl=len(loc[0])   
        index=h
        locl=loc
        w, h = temp.shape[::-1]
 
        
```

```python
for pt in zip(*locl[::-1]):
    cv2.rectangle(img_bgr, pt, (pt[0] + w, pt[1] + h), (0,255,255), 2)
```
