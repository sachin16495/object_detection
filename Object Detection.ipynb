```python
img_bgr = cv2.imread('Samsung_Train_Image/Test/samsung4.jpeg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

#img2 = img.copy()
#if img is None:

#img2 = img.copy()
#print img_bgr
```

```python
template = cv2.imread('Samsung_Train_Image/Train/trainimg12.jpeg',0)
#print template
w, h = template.shape[::-1]
```

```python
res=cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
sres=cv2.matchTemplate(img_gray,template,cv2.TM_SQDIFF)
```

```python
print np.sum(res)
print res.size
```

```python
threshold = 0.8
loc = np.where( res >= threshold)
loc
```

```python
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_bgr, pt, (pt[0] + w, pt[1] + h), (0,255,255), 2)
```
